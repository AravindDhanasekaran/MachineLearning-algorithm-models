```python
#getting the row and column size 
data.shape
```

```python
#we are converting our data to into array because we want tokenize each letter in the word to classify the strenght of password
password=np.array(data)
password
```

```python
x=[label[0] for label in password]#independent features
y=[label[1] for label in password]#dependent features
```

```python
x
```

```python
y
```

```python
x.shape`
```

```python
#this function is used to split the letters of the password
#we are spliting the letter because we can tokenize the letter by letter using tfidf vectorizer
def sepchar(inputs):
    characters=[]
    for i in inputs:
        characters.append(i)
    return characters
```

```python
#importing td-idf vectorizer for tokenizing
from sklearn.feature_extraction.text import TfidfVectorizer
vectors=TfidfVectorizer(tokenizer=sepchar)
x=vectors.fit_transform(x)
```

```python
#Getting the shape of the tokenized x
x.shape
```

```python
vectors.vocabulary_
```

```python
data.iloc[0,0]
```

```python
# to check how a single password is converted into features 
features=vectors.get_feature_names()
first=x[0]
df=pd.DataFrame(first.T.todense(),index=features,columns=["tfidf"])
df.sort_values(by=["tfidf"],ascending=False)
```

```python
#importing libaries to perform logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
#spliting trainset and training set
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
logis=LogisticRegression()
logis.fit(xtrain,ytrain)
```

```python
print(logis.score(xtest,ytest))
```

```python

X_predict=np.array(["%@123abcd"])
X_predict=vectors.transform(X_predict)
y_pred=logis.predict(X_predict)
print(y_pred)
```
