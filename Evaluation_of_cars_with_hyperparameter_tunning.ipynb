#### Importing libraries

```python
#importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV
```

#### Loading the dataset

```python
data=pd.read_csv("car.csv")
```

```python
data.head()
```

```python
data.tail()
```

#### Converting the categorical variable

```python
from sklearn import linear_model, preprocessing
```

```python
label_encoding=preprocessing.LabelEncoder()
```

```python
buying=label_encoding.fit_transform(list(data["buying"]))
maint=label_encoding.fit_transform(list(data["maint"]))
door=label_encoding.fit_transform(list(data["door"]))
persons=label_encoding.fit_transform(list(data["persons"]))
lug_boot=label_encoding.fit_transform(list(data["lug_boot"]))
safety=label_encoding.fit_transform(list(data["safety"]))
cls=label_encoding.fit_transform(list(data["class"]))
```

#### Spliting features into independent and dependent  features

```python
x=list(zip(buying,maint,door,persons,lug_boot,safety))
y=list(cls)
x
```

#### Spliting the train and test sets 

```python
from sklearn.model_selection import train_test_split 
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1)
```

#### Creating a Classification model(K-Nearest neigbhour) and Deploying it

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=3)
```

```python
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)
```

#### Measuring the performance of the model

```python
print(classification_report(ytest, ypred,))
```

#### Parameter tunning of our KNN model

```python
### List Hyperparameters that we want to tune.
leaf_size = list(range(1,50))
n_neighbors = list(range(1,30))
p=[1,2]
#Con`vert to dictionary
#dict = {'max_features': [4,5,7,9,11], 'max_leaf_nodes' : [5,7,9,13,19]}
hyperparameters={'leaf_size':leaf_size, 'n_neighbors':n_neighbors, 'p':p}
#Create new KNN object
knn_2 = KNeighborsClassifier()
#Use GridSearch
clf = GridSearchCV(knn_2, hyperparameters, cv=10)
#Fit the model
best_model = clf.fit(x,y)
#Print The value of best Hyperparameters
print('Best leaf_size:', best_model.best_estimator_.get_params()['leaf_size'])
print('Best p:', best_model.best_estimator_.get_params()['p'])
print('Best n_neighbors:', best_model.best_estimator_.get_params()['n_neighbors'])
```

#### Creating KNN model based on the best parameters

```python
model1=KNeighborsClassifier(n_neighbors=5,leaf_size=4,p=1)
model1.fit(xtrain,ytrain)
ypred1=model1.predict(xtest)
```

#### Perfromance of the tunned model

```python
print(classification_report(ytest, ypred1,))
```
